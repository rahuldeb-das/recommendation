```python
def reorder(train):
    train.index = train['customerID'].tolist()
    train.drop('customerID', axis=1, inplace=True)
    train.drop('Gender', axis=1, inplace=True)
    train.drop('State', axis=1, inplace=True)
    train.drop('transactionDate', axis=1, inplace=True)
    train.drop('target', axis=1, inplace=True)
    #train.drop('storecode', axis=1, inplace=True)
    #train.drop('transactionnumberbytill', axis=1, inplace=True)
    train = train.reindex(users)
    return train
```

```python
def svd(train, trueMat, k):
    utilMat = np.array(train)

    mask = np.isnan(utilMat)
    masked_arr=np.ma.masked_array(utilMat, mask)
    item_means=np.mean(masked_arr, axis=0)
    utilMat = masked_arr.filled(item_means)

    x = np.tile(item_means, (utilMat.shape[0],1))

    utilMat = utilMat - x

    print(utilMat)

    U, s, V=np.linalg.svd(utilMat, full_matrices=False)
    s=np.diag(s)
    s=s[0:k,0:k]
    U=U[:,0:k]
    V=V[0:k,:]

    s_root=sqrtm(s)

    Usk=np.dot(U,s_root)
    skV=np.dot(s_root,V)
    UsV = np.dot(Usk, skV)

    UsV = UsV + x

    UsV = np.ma.masked_where(trueMat==1,UsV).filled(fill_value=-999)

    #print(UsV)
    print("svd done")
    return UsV
```

```python
def max_items(UsV,x,j):
    out = []

    for xx in x:
        if UsV[j,xx]>0.001: # setting a threshold
            out.append(items[xx])
        
    if len(out) < 20:
        for j in range(20 - len(out)):
            out.append("None")
        

    return out


```

```python
def recommend(test):

    UsV = (5*svd(train=train,trueMat=trueMat,k=16) + 5*svd(train=train, trueMat=trueMat, k=16))/10
    #UsV = svd(train=train,trueMat=trueMat,k=8)/2
    pred = []
    testusers = test['customerID'].tolist()

    n = 20
    for user in testusers[:]:
        if user not in u:
            pred.append("300776411,300776410,108037568,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None")
            #pred.append("1000256435,300785148,1000256370,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None,None")
            test['products'] = pd.Series(pred)
        else:
            j = u[user]
            p = max_items(UsV, UsV[j,:].argsort()[-n:][::-1], j)
            pred.append(",".join(p))
            print(len(p))
            test['products'] = pd.Series(pred)
    #test['products'] = pred
    test.to_csv('final_svd_k16.csv', index=False)
```
